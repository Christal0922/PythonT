```python
# 
def get_sum(num):
    s=0
    for i in range(1,num+1):
        s+=i
    print(f'1{num}{s}')
get_sum(10)
```

```python
# 
def fun(*para):
    print(type(para))
    for item in para:
        print(item)
fun(10,20,30)
# 
fun([12,23,34,56])
fun(*[12,23,34,56])
```

```python
# 
def fun2(**kwpara):
    print(type(kwpara))
    for key,value in kwpara.items():
        print(key,'--->',value)
fun2(name='Tina',age=18,height=150)
d={'name':'Tina','age':18,'height':150}
fun2(**d)
```

```python
def fac(n):
    if n==1:
        return 1
    else:
        return n*fac(n-1)
print(fac(6))
```

```python
# 
# 1 1 2 3 5 8 13
def fib(n):
    if n==1:
        return 1
    if n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)
print(fib(6))
for i in range(1,7):
    print(fib(i),end=' ')
```
