# 格式化字符串
+ 格式化字符串的方式



```python
# 使用占位符进行格式化
name='马冬梅'
age=18
score=98.5
print('姓名：%s,年龄: %d, 成绩：%f' % (name,age,score))
print('姓名：%s,年龄: %d, 成绩：%.2f' % (name,age,score)) #保留两位小数
```

```python
# f-string 格式化字符串
print(f'姓名：{name},年龄:{age}, 成绩：{score}')
```

```python
# format
print('姓名：{0},年龄:{1}, 成绩：{2}'.format(name,age,score))

# 对齐
s='helloworld'
print('{0:*<20}'.format(s)) # 字符串显示宽度20，左对齐，空白部分使用*填充
print('{0:*>20}'.format(s)) # 右对齐
print('{0:*^20}'.format(s)) # 居中对齐
print(s.center(20,'*')) # 实现居中对齐

# 千位分隔符(只适用于整数和浮点数)
print('{0:,}'.format(98765747389))
print('{0:,}'.format(98765747389.976))

# 浮点数小数部分的精度
print('{0:.2f}'.format(3.1415926))
# 或者字符串类型的最大显示长度
print('{0:.5}'.format('helloworld'))
```

```python
# 整数类型
a=425
print('{0:b},{0:c},{0:d},{0:o},{0:x},{0:X}'.format(a))
# 二进制，UNICODE符号, 十进制，八进制，16进制小写，16进制大写

# 浮点数类型
b=3.1415926
print('{0:.2f},{0:.2E},{0:.2e},{0:.2%}'.format(b))
```

# 字符串的编码与解码
+ 字符串编码


## python中两种常用的字符串类型


## 字符串的编码


```python
# str --> byte
s='伟大的中国梦'
scode_gbk=s.encode('gbk')
print(scode_gbk)
scode_utf8=s.encode('utf-8') # 默认为utf-8
print(scode_utf8)

# byte --> str
print(bytes.decode(scode_gbk,'gbk'))
print(bytes.decode(scode_utf8,'utf-8'))
```

# 数据的验证

## 数据的验证
+ 对用户输入的数据进行‘合法’性验证


```python
# 所有字符都是数字（十进制的阿拉伯数字）
print('123'.isdigit()) # True
print('一二三'.isdigit()) # False
print('0b1001'.isdigit()) # False
print('ⅠⅡⅢ'.isdigit()) # False
print('-'*30)
# 所有字符都是数字（阿拉伯数字、罗马数字）
print('123'.isnumeric()) # True
print('一二三'.isnumeric()) # True
print('0b1001'.isnumeric()) # False
print('ⅠⅡⅢ'.isnumeric()) # True
print('壹贰叁'.isnumeric()) # True
```

```python
# 所有都是字母（英文、中文）
print('hello你好'.isalpha()) # True
print('hello你好123'.isalpha()) # False
print('hello你好一二三'.isalpha()) # True
print('hello你好ⅠⅡⅢ'.isalpha()) # False
print('-'*50)
# 所有都是字母+数字
print('hello你好123'.isalnum()) # True
print('hello你好123...'.isalnum()) # False
```

```python
# 所有字符都是小写吗？
print('Helloworld'.islower()) # False
print('helloworld'.islower()) # True
print('hello你好'.islower()) # True
print('-'*50)

# 所有字符都是大写吗？
print('Helloworld'.isupper()) # False
print('HELLOWORLD'.isupper()) # True
print('HELLO你好'.isupper()) # True
print('-'*50)

# 首字母大写
print('HelloWorld'.istitle()) # False
print('Helloworld'.istitle()) # True
print('-'*50)

# 是否都是空白字符
print('\t'.isspace()) # True
print('\n'.isspace()) # True
print('   '.isspace()) # True
```

## 字符串的去重

```python
# 字符串去重
s='helloworldhelloworldhelloworld'
# not in
new_s=''
for item in s:
    if item not in new_s:
        new_s+=item
print(new_s)

# 使用索引+not in 
new_s2=''
for i in range(len(s)):
    if s[i] not in new_s2:
        new_s2+=s[i]
print(new_s2)
# 通过集合去重+列表排序
new_s3=set(s)
lst=list(new_s3)
lst.sort(key=s.index)
print(''.join(lst))
```

```python
# 列表元素的去重
lst=['金星','木星','水星','火星','土星','金星','木星','水星','火星','土星']
# 法一： 遍历+ not in
new_lst=[]
for item in lst:
    if item not in new_lst:
        new_lst.append(item)
print(new_lst)

# 法二： 遍历+ not in + 索引
new_lst2=[]
for i in range(len(lst)):
    if lst[i] not in new_lst2:
        new_lst2.append(lst[i])
print(new_lst2)
print(','.join(new_lst2))

# 法三：集合+排序
new_s3=list(set(lst))
new_s3.sort(key=lst.index)
print(','.join(new_s3))
```

# 内置模块re的使用


## re.match()
从头开始匹配

```python
import re
pattern=r'\d\.\d+' # r表示元字符，转义字符不起作用,\d匹配任意十进制数,
#\.任意字符，+ 一次或多次
s='I study Python3.10 every day'
s2='3.10Python'
match=re.match(pattern,s2,re.I)
print(match)
print('匹配值的起始位置：',match.start())
print('匹配值的结束位置：',match.end())
print('匹配区间的位置：',match.span())
print('待匹配的字符串：',match.string)
print('匹配的数据：',match.group())
```

## re.search()


```python
import re
pattern=r'\d\.\d+'
s='Python3.10'
s2='4.10Python'
s3='Python'
match=re.search(pattern,s)
match2=re.search(pattern,s2)
match3=re.search(pattern,s3)
print(match)
print(match2)
print(match3)
```

```python
import re
pattern=r'\d\.\d+'
s='Python3.10,python2.10'
s2='4.10Python'
s3='Python'
lst=re.findall(pattern,s)
lst2=re.findall(pattern,s2)
lst3=re.findall(pattern,s3)
print(lst)
print(lst2)
print(lst3)
```
