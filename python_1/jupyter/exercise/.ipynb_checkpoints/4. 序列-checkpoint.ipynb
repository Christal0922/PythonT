# 序列和列表
序列[start:end:step]

## 切片

```python
s='HelloWorld'
s1=s[0:5:1]
print(s1)
# 省略开始位置
print(s[:5:1])
# 省略结束位置
print(s[0::1])
# 省略结束位置和步长
print(s[5:])
# 省略开始和结束，只写步长
print(s[::2])
# 步长可以为负数
print(s[::-1])
```

## 序列的相关操作

```python
# 序列的相加
s='Hello'
s2='World'
print(s+s2)
# 注意事项 +左右的数据类型相同，序列中元素的数据类型可以不同
```

```python
# 序列的相乘
s='Hello'
print(s*5)
```

```python
# 序列的相关操作符
s='helloworld'
print('e在helloworld中存在吗？',('e' in s))
print('e不在helloworld中存在吗？',('e' not in s))
print('len():',len(s))
print('max():',max(s))
print('min():',min(s))

# 序列对象的方法，使用序列的名称，打点调用
print('s.index()',s.index('w'))
print('s.count()',s.count('o'))
```

# 列表

## 列表的创建和删除

```python
# 列表的创建
## 直接使用[]创建
lst=['hello','world',99.8,100]
## 使用list()函数创建
lst2=list('helloworld')
lst3=list(range(1,10,2)) #从1开始，到10结束（不包含10），步长为2
# 列表是序列中的一种，对序列操作的运算符，操作符，函数均可使用
print(lst+lst2+lst3)
print(lst*3)
print(lst2.count('o'))
print(lst2.index('o'))
```

```python
# 列表的删除
lst4=[10,20,30]
print(lst4)
del lst4
```

## 列表的遍历

```python
lst=['hello','world','python','php']
# 使用for
for item in lst:
    print(item)
# 使用for循环，range()，len()
for i in range(len(lst)):
    print(i,'-->',lst[i])
# 使用for enumerate
for index,item in enumerate(lst):
    print(index,item)
for index,item in enumerate(lst,10): # 序号从10开始
    print(index,item)
```

## 列表操作

列表中实际存储的是对象的内存地址

```python
# 列表元素的增加
lst=['hello','world','python']
print(lst)
lst.append('sql')
lst.insert(1,100)
```

```python
# 列表元素的删除
lst=['hello','world','python']
lst.remove('world')
print(lst.pop(1))
lst.clear()
```

```python
# 列表元素的改变
lst=['hello','world','python']
lst.reverse()

# 列表的拷贝
new_list=lst.copy()
```

```python
# 列表元素的修改
lst=['hello','world','python']
lst[1]='mysql'
```

## 列表的排序
sort()  
sorted() 原列表元素的顺序不变，排序后产生一个新的列表对象

```python
# sort
lst=[4,56,3,78,40,56,89]
print('原列表',lst)

lst.sort()
print('升序：',lst)
lst.sort(reverse=True)
print('降序：',lst)
```

```python
lst2=['banana','apple','Cat','Orange']
#升序，先排大写后排小写
lst2.sort()
print('升序：',lst2)
# 忽略大小写进行比较
lst2.sort(key=str.lower)
print(lst2)

new_lst2=sorted(lst2,key=str.lower)
print(new_lst2)
```

```python
# sorted
lst=[4,56,3,78,40,56,89]
asc_lst=sorted(lst)
desc_lst=sorted(lst,reverse=True)
print(lst,asc_lst,desc_lst)
```

```python
# 列表生成式
lst=[item for item in range(1,11)]
print(lst)
lst=[item*item for item in range(1,11)]
print(lst)
```

```python
# 生成随机数
import random
lst=[random.randint(1,100) for _ in range(10)]
print(lst)
```

```python
# 从列表中选择符合条件的元素组成新的列表
lst=[i for i in range(10) if i%2==0]
print(lst)
```

## 二维列表

```python
lst=[
    ['城市','环比','同比'],
    ['北京',102,103],
    ['上海',104,584],
    ['深圳',100,39]
]
#
for row in lst: #行
    for item in row: #列
        print(item,end='\t')
    print()
    
```

```python
# 列表生成式生成四行五列
lst2=[[j for j in range(5)] for i in range(4)]
print(lst2)
```
