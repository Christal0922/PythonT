```python
class Person(object): # Personobject
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def info(self):
        print('{0}{1}'.format(self.name,self.age))
```

```python
# 
class Student(Person):
    def __init__(self,name,age,score):
        super().__init__(name,age)
        self.score=score
# 
stu=Student('Jack',20,'1001')
stu.info()
```

## 

```python
# 
class Student(Person):
    def __init__(self,name,age,score):
        super().__init__(name,age)
        self.score=score
    def info(self):
        super().info()
        print('{0}'.format(self.score))
# 
stu=Student('Jack',20,'1001')
stu.info()
```

```python
class Animal(object):
    def eat(self):
        print('')


class Dog(Animal):
    def eat(self):
        print('')


class Cat(Animal):
    def eat(self):
        print('')


class Person(Animal):
    def eat(self):
        print('')


def fun(animal):
    animal.eat()


fun(Dog())
fun(Cat())
fun(Person())
```
