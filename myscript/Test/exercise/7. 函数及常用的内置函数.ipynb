# 函数的定义及调用



```python
# 计算累加和
def get_sum(num):
    s=0
    for i in range(1,num+1):
        s+=i
    print(f'1到{num}之间的累加和为：{s}')
get_sum(10)
```

## 参数的类型




```python
def happy_birthday(name,age):
    print('祝{0}{1}岁生日快乐!'.format(name,str(age)))
happy_birthday('Tina',18)
happy_birthday(name='Tina',age=18)
```

```python
# 可变参数
def fun(*para):
    print(type(para))
    for item in para:
        print(item)
fun(10,20,30)
# 调用时参数前加个星，会对列表进行解包
fun([12,23,34,56])
fun(*[12,23,34,56])
```

```python
# 个数可变的关键字参数
def fun2(**kwpara):
    print(type(kwpara))
    for key,value in kwpara.items():
        print(key,'--->',value)
fun2(name='Tina',age=18,height=150)
d={'name':'Tina','age':18,'height':150}
fun2(**d)
```

# 函数的返回值
+ 如果函数没有返回值，return可以省略
+ 如果函数的返回值为1个，直接返回类型
+ 如果函数的返回值为多个，返回的结果是元组

```python
# 计算奇数、偶数
def fun(num):
    odd=[]
    even=[]
    for i in num:
        if i%2:
            odd.append(i)
        else:
            even.append(i)
    return odd,even
print(fun([10,29,34,23,44,53,55]))
```

# 递归函数




```python
def fac(n):
    if n==1:
        return 1
    else:
        return n*fac(n-1)
print(fac(6))
```

```python
# 菲波那契数列
# 1 1 2 3 5 8 13
def fib(n):
    if n==1:
        return 1
    if n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)
print(fib(6))
for i in range(1,7):
    print(fib(i),end=' ')
```
