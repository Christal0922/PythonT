# 元组


## 元组的创建和删除

```python
# 直接使用（）创建元组
t=('hello',[10,20,30],'python','world')
print(t)

# 使用tuple()创建元组
t=tuple('helloworld')
print(t)

t=tuple(range(1,10))
print(t)

t=tuple([10,20,30,40])
print(t)

```

```python
# 元组的相关操作
t=tuple(range(1,11))
print('10在元组中是否存在',(10 in t))
print('len:',len(t))
print('t.index:',t.index(10)) # 数字10的位置
print('t.count:',t.count(10)) # 数字10出现的次数

x=(10)
print(x,type(x))

y=(10,)
print(y,type(y))
# 元组的删除
del y
```

## 元组的访问与遍历
生成器遍历之后，再想重新遍历必须重新创建一个生成器对象，因为遍历之后原生成器对象已经不存在了

```python
# 元组支持切片操作
t=('python','hello','world')
print(t[0])
print(t[0:3:2])
```

```python
# 元组的遍历
t=('python','hello','world')
for item in t:
    print(item)

# for+range()+len
for index in range(len(t)):
    print(index,t[index])

# enumerate()
for index,item in enumerate(t):
    print(index,'-->',item)
for index,item in enumerate(t,1): # 从1开始
    print(index,'-->',item)
```

## 元组生成式

```python
t=(i for i in range(1,11)) # 结果是一个生成器对象
t=tuple(t)
print(t)
```

```python
# for 
t=(i for i in range(1,5))
for item in t:
    print(item)

# __next__()方法,取出
t=(i for i in range(1,5))
print(t.__next__())
print(t.__next__())
t=tuple(t)
print(t)
```

## 字典的创建和删除

```python
# 使用{}直接创建
d={10:'cat',20:'dog',30:'pet',40:'rabbit'}

# zip函数
lst1=[10,20,30,40]
lst2=['cat','dog','pet','rabbit']
zipobj=zip(lst1,lst2) # 映射函数的结果是一个zip对象
print(list(zipobj))
d=dict(d)
print(d)

# 使用参数创建字典
d=dict(cat=10,dog=20,pet=30,rabbit=40) # 注意事项，参数相当于变量，变量的名字不加引号
print(d)

# 使用元组
t=(10,20,30)
print({t:10})
```

## 字典元素的访问和遍历


```python
d={'cat':10,'dog':20,'pet':30,'rabbit':40}
print(d)
# 使用[key]
print(d['dog'])
# 使用d.get(key)
print(d.get('tiger','不存在'))
```

```python
d={'cat':10,'dog':20,'pet':30,'rabbit':40}
for item in d.items():
    print(item)
    
for key,value in d.items():
    print(key,value)
```

## 字典的相关操作


```python
# 字典的添加
d={10:'cat',20:'dog',30:'pet',40:'rabbit'}
d[50]='tiger'
print(d)
print(d[50])

# 获取字典中所有的key
keys=d.keys() # d.keys()结果是dict_keys, python中的一种内部数据结构，专用于表示字典的key
# 如果希望更好的显示数据，可以使用list或者tuple转成相应的数据类型
print(keys)
print(list(keys))
print(tuple(keys))

# 获取字典中所有的value
values=d.values
print(values)
print(list(values))
print(tuple(values))

# 字典遍历时用到的一个方法items
items=d.items()
print(items)
print(list(items))
```

## 字典生成式
+ 使用指定范围的数作key  
    d={key:value for item in range}
+ 使用映射函数生成字典  
    d={key:value for key,value in zip(lst1,lst2)}

```python
import random
d={item:random.randint(1,100) for item in range(4)}
print(d)
```

```python
lst1=[1001,1002,1003]
lst2=['Mark','Lina','Tie']
d={key:value for key,value in zip(lst1,lst2)}
print(d)
```

# 集合
+ 集合的特点


## 集合的创建和删除
+ 使用{}直接创建集合  
    s={元素1，元素2，......,元素N}
+ 使用内置函数set()创建集合  
    s=set(可迭代对象）
+ 集合的删除  
    del 集合名

```python
s={10,20,30,40}
print(s)
s={}
print(type(s))

s=set()
print(type(s),bool(s))

```

```python
s1=set('helloworld')
s2=set([10,20,30])
s3=set(range(1,10))
print(s1,s2,s3)
```

```python
del s1,s2,s3
```

## 集合的操作符


```python
A={10,20,30,40,50}
B={30,50,88,76,20}

#
print(A&B) #交集
print(A|B) #并集
print(A-B) # 差集
print(A^B) # 补集
```

## 集合的相关操作


```python
s={10,20,30}
s.add(100)
print(s)
s.remove(20)
print(s)
s.clear()
print(s,'空集合的布尔值',bool(s))
```

```python
# 遍历集合
s={10,20,30}
for item in s:
    print(item)
for index,item in enumerate(s,10): # 元素的序号，可以自定义
    print(index,'-->',item)

```

```python
s={i for i in range(10)}
print(s)
s={i for i in range(10) if i%2==0}
print(s)
```

# 组合数据类型的总结
+ 列表、元素、字典和集合的区别
|数据类型|序列类型|是否重复|是否有序|字义符号|
|:--:|:--:|:--:|:--:|:--:|
|元组list|可变序列|可重复|有序|[]|
|元组tuple|不可变序列|可重复|有序|()|
|字典dict|可变序列|key不可重复  value可重复|无序|{key:value}|
|集合set|可变序列|不可重复|无序|{}|
