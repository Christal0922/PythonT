```python
# 
t=('python','hello','world')
for item in t:
    print(item)

# for+range()+len
for index in range(len(t)):
    print(index,t[index])

# enumerate()
for index,item in enumerate(t):
    print(index,'-->',item)
for index,item in enumerate(t,1): # 1
    print(index,'-->',item)
```

## 


```python
# 
d={10:'cat',20:'dog',30:'pet',40:'rabbit'}
d[50]='tiger'
print(d)
print(d[50])

# key
keys=d.keys() # d.keys()dict_keys, pythonkey
# listtuple
print(keys)
print(list(keys))
print(tuple(keys))

# value
values=d.values
print(values)
print(list(values))
print(tuple(values))

# items
items=d.items()
print(items)
print(list(items))
```
