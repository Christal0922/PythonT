```python
lst=[88,89,90,0,99]
print(':',lst)
lst2=[]
for item in lst:
        if item>= 10:
            lst2.append('19'+str(item))
        else:
            lst2.append('200'+str(item))
print(lst2)
        
```

```python
lst=[88,89,90,0,99]
print(':',lst)
for index in range(len(lst)):
    if lst[index]==0:
        lst[index]='200'+str(lst[index])
    else:
        lst[index]='19'+str(lst[index])
print('',lst)
```

```python
lst=[88,89,90,0,99]
print(':',lst)
for index,value in enumerate(lst):
    if value==0:
        lst[index]='200'+str(value)
    else:
        lst[index]='19'+str(value)
print('',lst)
```
