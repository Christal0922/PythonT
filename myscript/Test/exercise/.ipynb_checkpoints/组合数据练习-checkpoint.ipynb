# “千年虫”是什么虫


```python
lst=[88,89,90,0,99]
print('原列表:',lst)
lst2=[]
for item in lst:
        if item>= 10:
            lst2.append('19'+str(item))
        else:
            lst2.append('200'+str(item))
print(lst2)
        
```

```python
lst=[88,89,90,0,99]
print('原列表:',lst)
for index in range(len(lst)):
    if lst[index]==0:
        lst[index]='200'+str(lst[index])
    else:
        lst[index]='19'+str(lst[index])
print('更改后的列表：',lst)
```

```python
lst=[88,89,90,0,99]
print('原列表:',lst)
for index,value in enumerate(lst):
    if value==0:
        lst[index]='200'+str(value)
    else:
        lst[index]='19'+str(value)
print('更改后的列表：',lst)
```

# 模拟京东购物流程


```python
lst=[]
for i in range(5):
    goods=input('请输入商品的编号和商品的名称进行商品入库，每次只能输入一件商品')
    lst.append(goods)
for item in lst:
    print(item)
```

```python
# 购物车
cart=[]
while True:
    flag=False # 
    num=input('请输入要购买的商品编号：')
    for item in lst:
        if num==item[0:4]:
            flag=True
            cart.append(item)
            print('商品已成功添加到购物车')
            break
    if flag==False and num!='q':
        print('该商品不存在！')
    if num=='q':
        break
print('您购物车里已选择的商品为：')
#反向
cart.reverse()
for item in cart:
    print(item)
```

# 模拟12306火车票的订票流程


```python
# 创建字典用于存储车票的信息
dict_ticket={
    'G1569':['北京南-天津南','18:05','18.39','00:34'],
    'G1567':['北京南-天津南','18:15','18.49','00:34'],
    'G8917':['北京南-天津南','18:20','19:19','00:59'],
    'G203':['北京南-天津南','18:35','19:09','00:34']
}
# 遍历字典中的元素
print('车次\t出发站-到达站\t出发时间\t到达时间\t历时')
for key in dict_ticket.keys():
    print(key,end='\t')
    for item in dict_ticket.get(key):
        print(item,end='\t')
    print()
        
```

```python
train_no=input('请输入要购买的车次：')
info=dict_ticket.get(train_no,'车次不存在')
if info!='车次不存在':
    person=input('请输入乘车人，如果是多人使用请用逗号分隔')
    s=info[0]+' '+info[1]+'开'
    print('您已购买了'+train_no+'次列车,'+s+
      '请'+person+',尽快换取纸质车票。【铁路客服】')
```

```python
for key,value in dict_ticket.items():
    print(key,end='\t')
    for i in dict_ticket[key]:
        print(i,end='\t')
    print()
```
